1. Explain the different types of data (qualitative and quantitative) and provide examples of each. Discuss nominal, ordinal, interval, and ratio scales.**  
**Answer:**  
- **Qualitative Data (Categorical Data):** Descriptive and non-numerical.  
  - **Examples:** Eye color, gender, car brands.  
- **Quantitative Data (Numerical Data):** Represents counts or measurements.  
  - **Examples:** Height, exam scores, temperature.  

 **Scales of Measurement:
- **Nominal:** Categories without order (e.g., blood type: A, B, AB, O).  
- **Ordinal:** Ordered categories without equal differences (e.g., customer satisfaction: satisfied, neutral, dissatisfied).  
- **Interval:** Numerical, ordered, equal intervals but no true zero (e.g., temperature in Celsius).  
- **Ratio:** Numerical, ordered, equal intervals with a true zero (e.g., weight, height).

### **2. What are the measures of central tendency, and when should you use each? Discuss the mean, median, and mode with examples and situations where each is appropriate.**  
**Answer:**  
- **Mean:** The average value, used for symmetrical distributions (e.g., test scores).  
- **Median:** The middle value, useful for skewed data (e.g., house prices).  
- **Mode:** The most frequent value, used for categorical data (e.g., most common shirt size).  



### **3. Explain the concept of dispersion. How do variance and standard deviation measure the spread of data?**  
**Answer:**  
- **Dispersion** measures how spread out data points are.  
- **Variance:** The average squared deviation from the mean.  
- **Standard Deviation:** The square root of variance, showing spread in the same units as data.  



### **4. What is a box plot, and what can it tell you about the distribution of data?**  
**Answer:**  
A **box plot** visualizes data distribution using quartiles and outliers. It shows:  
- **Median** (central value).  
- **Interquartile Range (IQR)** (middle 50% of data).  
- **Minimum & Maximum values** (excluding outliers).  
- **Outliers** (extreme data points).  
It helps in identifying skewness and data spread.  



### **5. Discuss the role of random sampling in making inferences about populations.**  
**Answer:**  
- **Random sampling** ensures a representative subset of a population.  
- Reduces **bias**, making statistical inferences more reliable.  
- Used in hypothesis testing and confidence intervals to generalize findings.  



### **6. Explain the concept of skewness and its types. How does skewness affect the interpretation of data?**  
**Answer:**  
- **Skewness** measures asymmetry in a dataset.  
  - **Positive Skew:** Tail on the right (e.g., income distribution).  
  - **Negative Skew:** Tail on the left (e.g., exam scores with many high values).  
- Skewness affects the relationship between **mean, median, and mode**, influencing data interpretation.  



### **7. What is the interquartile range (IQR), and how is it used to detect outliers?**  
**Answer:**  
- **Interquartile Range (IQR) = Q3 - Q1**, representing the middle 50% of data.  
- **Outliers** are values **below Q1 - 1.5(IQR) or above Q3 + 1.5(IQR)**.  
- Used to identify extreme values in datasets.  



### **8. Discuss the conditions under which the binomial distribution is used.**  
**Answer:**  
A **binomial distribution** is used when:  
- There are **fixed** trials (n).  
- Each trial has **two possible outcomes** (success/failure).  
- The probability of success (**p**) is constant.  
- Trials are **independent** (one outcome doesn’t affect the next).  

Example: Flipping a coin 10 times and counting heads.  



### **9. Explain the properties of the normal distribution and the empirical rule (68-95-99.7 rule).**  
**Answer:**  
- **Normal Distribution:** Bell-shaped, symmetric, mean = median = mode.  
- **Empirical Rule:**  
  - **68%** of data falls within **1 standard deviation** of the mean.  
  - **95%** within **2 standard deviations**.  
  - **99.7%** within **3 standard deviations**.  

Used in probability and statistics for real-world data approximations.  



### **10. Provide a real-life example of a Poisson process and calculate the probability for a specific event.**  
**Answer:**  
Example: The number of customer calls received in an hour at a call center.  

If the average calls per hour = **λ = 5**, the probability of getting exactly **k = 3** calls in an hour:  
\[
P(X = k) = \frac{e^{-λ} λ^k}{k!}
\]
\[
P(X = 3) = \frac{e^{-5} 5^3}{3!} = \frac{0.0067 \times 125}{6} = 0.1404
\]  
So, the probability of 3 calls in an hour is **14.04%**.

### **12. Provide an example dataset, calculate both covariance and correlation, and interpret the results.**  
**Answer:**  
Dataset:  
\[
X = [2, 4, 6, 8], \quad Y = [5, 10, 15, 20]
\]  

#### **Step 1: Compute Covariance**  
\[
Cov(X, Y) = \frac{\sum (X_i - \bar{X})(Y_i - \bar{Y})}{n-1}
\]
Result: **Cov(X, Y) = 10** (positive relationship).  

#### **Step 2: Compute Correlation**  
\[
r = \frac{Cov(X, Y)}{\sigma_X \sigma_Y}
\]
Result: **r = 1** (perfect positive correlation).  

**Interpretation:** X and Y have a strong **direct linear relationship**.
